```python
import pandas as pd

# サンプルデータ（CSVから読み込む場合は pd.read_csv を使ってね）
df = pd.DataFrame({
    'latitude': [35.6895, 34.6937, 43.0642, 33.5902, 38.2688],
    'longitude': [139.6917, 135.5023, 141.3469, 130.4017, 140.8721],
    'value': [80, 60, 40, 50, 30]
})
```

```python
import folium
from folium.plugins import HeatMap

# 地図の初期設定（日本の中心）
m = folium.Map(location=[36.2048, 138.2529], zoom_start=5)

# DataFrameから [緯度, 経度, 値] のリストを作成
heat_data = df[['latitude', 'longitude', 'value']].values.tolist()

# ヒートマップを追加
HeatMap(heat_data).add_to(m)

# Jupyter Notebook上で表示
m

```
